# 6.FLIPKART 100 SUNGLASSES

```python
url_6='https://www.flipkart.com/'
driver.get(url_6)
```

```python
cross = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
cross.click()
time.sleep(5)
```

```python
search = driver.find_element_by_xpath('//div[@class="_3OO5Xc"]//input')
search.clear()
search.send_keys('sunglasses')

search_button = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_button.click()
time.sleep(5)

brand = []
description = []
price = []
discount = []


for a in range (3):
    brand_tags = driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
    desc_tags = driver.find_elements_by_xpath('//div[@class="_2B099V"]//a')
    price_tags = driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
    discount_tags = driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]//span')

    for i in brand_tags:
        brand.append(i.text)
    
    for i in desc_tags:
        description.append(i.text)
    
    for i in price_tags:
        price.append(i.text)
    
    for i in discount_tags:
        discount.append(i.text)

    if (a < 1):
        nxt = driver.find_element_by_xpath('//a[@class="_1LKTO3"]')
        nxt.click()
        time.sleep(5)
    else:
        nxt = driver.find_element_by_xpath('//a[@class="_1LKTO3"][2]')
        nxt.click()
        time.sleep(5)

desc = description[0::2]
```

```python
sunglasses = ps.DataFrame({})
sunglasses['Brand'] = brand
sunglasses['Description'] = desc
sunglasses['Price'] = price
sunglasses['Discount'] = discount
sunglasses = sunglasses[:100]
sunglasses
```

# 7. Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link

```python
url7 = 'https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace'
driver.get(url7)
all_r = driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]')
all_r.click()
time.sleep(3)
```

```python
rate = []
rev_sum = []
rev_full = []

for a in range (10):
    rate_tags = driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]//div[@class="row"]')
    rev_sum_tags = driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
    rev_full_tags = driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    read_more_tags = driver.find_elements_by_xpath('//span[@class="_1BWGvX"]')

    for i in rate_tags:
        rate.append(i.text)
    
    for i in rev_sum_tags:
        rev_sum.append(i.text)
    
    for i in read_more_tags:
        i.click()
    
    for i in rev_full_tags:
        rev_full.append(i.text.replace('\n', ' '))
    
    if (a < 1):
        nxt = driver.find_element_by_xpath('//a[@class="_1LKTO3"]')
        nxt.click()
            
    else:
        nxt = driver.find_element_by_xpath('//a[@class="_1LKTO3"][2]')
        nxt.click()
            
    time.sleep(3)

ra = rate[0::4]
```

```python
review = ps.DataFrame({})
review['Ratings'] = ra
review['Review Summary'] = rev_sum
review['Full Review'] = rev_full
review
```
